## 1. Introduction and Background

#### A. The Motivating Question

Athleticism has long been a heavily desired trait, if not outright requirement, for the vast majority of sports hopefuls. From badminton to rugby, "players" have become synonymous with "athletes," complete with training regimens, health checkups, diet adjustments, and near-constant gym photos. At the same time, sports history is littered with examples of surprising successes from the less-athletic (Wayne Gretzky, Steve Nash, and the like), and conversely, the drafting process for almost every sport includes numerous busts who demonstrated excellent jumping or running ability but simply failed to play the game well. Basketball stands as an exemplar of this principle, and every year, countless draft experts and scouts dedicate sizable swaths of time to predicting which players will maximize their physical gifts, or else compensate with some other superb skill. In this notebook, we will seek to answer that underlying, motivating question: ***how does initial athleticism translate into on-court production?***

#### B. How NBA Basketball Works

* **Basic Structure:** The foremost league for basketball is the National Basketball Association (NBA), which has 82 games in a season. Every game is made up of 4 quarters of 12 minutes each, totalling 48 minutes per game. There are 10 players on the court at any given time, 5 from the home team and 5 from the away team, though in-game substitutions happen very frequently to give players rest. The team with the ball is on offense and tries to score points by throwing the ball in the basket, while the other team plays defense and tries to stop them.
<p>
* **Positions**: Typically, players are categorized as either guards, forwards, or centers. Guards are usually smaller and help to set up the offense; forwards are typically taller than guards and may act as jack-of-all-trades; and centers are generally the tallest players, tasked with anchoring the defense and getting the most rebounds (i.e., recovering the ball after a missed shot). These positions and their corresponding roles are not set in stone, however, and today's athletes often play multiple positions.
<p>
* **Box Score Statistics:** Most events that happen in-game are recorded in the "box score," with 5 main statistical cateogries. Most of these statistics are expressed on a per-game basis.
    1. Points are awarded for made shots.
    2. Rebounds are credited to players who recover the ball after a missed shot.
    3. Assists are credited to players who pass the ball to a teammate who makes a shot soon after.
    4. Steals are credited to players who take the ball away from the other team.
    5. Blocks are credited to players who physically alter the trajectory of a ball after a shot.
<br><br>
* **The Point System:** Like in most other sports, the team that scores the most points wins, and there are three ways to score points, each with their own corresponding point values. All shots that are taken in the flow of the game (e.g., everything but free throws) are called field goals.
    1. Made free throws are worth one point each. These are undefended shots that are taken about 15 feet away from the basket, and they are typically awarded when a player is fouled on a shot attempt, or in other words, when the defender makes physical contact with the shooter's arm or hand, or pushes the shooter out of position when the shot is taken.
    2. All shots taken inside the 3-point line are worth 2 points each, if they are made. 
    3. All shots taken outside the 3-point line are worth 3 points each, if they are made.
<br><br>
* **The Draft:** Every year, a new generation of players is drafted into the league, in a process with 2 rounds and 60 picks overall. While top draft picks often become stars or at least solid starters, many have also become busts, failing to live up to their hype and produce on the court. On the other hand, some players who went undrafted eventually achieved great success in the NBA.
 <p>
* **The Draft Combine:** The NBA Draft Combine is a yearly, invitation-only event which allows college-age basketball players to showcase their skills and physical abilities. The combine typically consists of both measurements and athletic tests. A player's performance at the combine can dramatically impact their draft position; for instance, out-of-shape players could be drafted much lower than they expected. Not all players who enter the NBA attend the combine, and not all players who attend the combine are eventually drafted.

#### D. Explaining the Variables

In our final dataset, we will examine the following variables:
* *Identifying Variables*
    * `Player`: The player's name.
    * `Pk`: The draft position of the player (from 1-60). Teams usually try to take better players at higher picks.
    * `Year`: The year that the player participated in the combine.
    * `Pos`: The player's position. May be guard (G), forward (F), center (C), or some combination of the two (e.g., G-F or F-C).
    * `Born`: The player's birth country, abbreviated to two letters. Most players are from the US.
    * `College`: The player's college, if applicable. International-born players usually do not attend American colleges.
    * `Age`: The player's age at the time of the combine. Most players are college-age (e.g., between 18 and 22).
<p>
* *Physical Measurements*
    * `Height (No Shoes)`: How tall the player is without shoes (inches). Most players in the NBA are at least 6 feet tall, but very few are over 7.5 feet.
    * `Height (With Shoes)`: How tall the player is with shoes (inches). Nearly all NBA players measure their height with shoes.
    * `Wingspan`: The fingertip-to-fingertip length of the player's extended arms (inches). The average male's wingspan is 2 inches greater than his height.
    * `Standing reach`: The maximum height the player can reach with his extended arms while standing (inches).
    * `Hand (Length)`: According to [howtheyplay.com](https://howtheyplay.com/team-sports/14-NBA-Players-With-the-Most-Impressive-Hand-Sizes), "Hand length is measured from the base of the palm to the tip of the middle finger. Very few NBA players have hand lengths of 10 inches or longer. Examples include Andrew Nicholson, Jaren Jackson Jr, and Dexter Pittman. The average hand length for a male is around 7.4 inches."
    * `Hand (Width)`: According to the same source as above, "Hand [width] is measured from the tip of the thumb to the tip of the pinky finger with your hand outstretched...it’s extremely rare for even an NBA player to have a hand span of 12 inches or wider. Gregory Smith and Giannis Antetokounmpo are a few who can boast of such measurements. The average male’s hand span is about 8.5 inches."
    * `Weight`: The player's weight in pounds.
    * `Body Fat`: The player's body fat percentage. For athletes, this number is usually [13% or less](https://www.acefitness.org/education-and-resources/lifestyle/blog/112/what-are-the-guidelines-for-percentage-of-body-fat-loss).
<p>
* *Athletic Tests*
    * Note: According to [coachmac-basketball.com](http://www.coachmac-basketball.com/average-vertical-jump/), the average male's max vertical is about 20 inches. 30 inches is standard for competititve men's basketball players, while a handful can achieve about 40 inches. Standing vertical is almost always lower than max vertical, since running starts are not allowed; the benchmark for a good standing vertical in the NBA is about 30 inches.
    * `Vertical (Max)`: The maximum height that a player can jump given a running start, measured from the ground to the bottom of the shoes (inches).
    * `Vertical (Max Reach)`: The maximum height that a player can reach with his extended arms while jumping, given a running start (inches).
    * `Vertical (No Step)`: The maximum height that a player can jump from a standstill position, measured from ground to shoes (inches).
    * `Vertical (No Step Reach)`: The maximum height that a player can reach with his extended arms while jumping from a standstill position (inches).
    * `Agility`: According to NBA writer [Mark Rappaport](https://www.nba.com/sixers/news/140512-combine-primer?fbclid=IwAR08ZyALMvkzpaY34IV04SU1QoXWmfoeumbsN8Jof1rj2d55B1XNl6KAINg), "The lane agility drill is designed to measure a player’s lateral quickness and ability to change direction. Cones are placed along the four corners of the key, and players follow a square path around them, first clockwise and then counterclockwise, peddling forwards, backwards, and side to side in the process." Most players can usually finish the drill in 10-12 seconds; sub-10 second times are extremely rare.
    * `Sprint`: The same source as above notes that "the distance covered in this drill is actually 75 feet despite the fact that an NBA court is 94 feet in length, the 3/4 court sprint is the basketball equivalent to football’s 40-yard dash. The drill is primarily a measure of acceleration rather than top-line speed, and times typically fall between the low- to mid-three second range."
<p>
* *Career Statistics*
    * `From`: The year that the player first started playing in the NBA.
    * `To`: The year the player stopped playing in the NBA. For active players, this year should be 2020, but it could be 2019 for players who have been injured for the entirety of 2019 and are returning.
    * `G`: How many career games the player has played in. 500 games is a standard for most experienced NBA players (about 6 seasons).
    * `MPG`: Minutes per game. Starters usually play about 20-40 minutes per game (with 40 minutes being at the very high end), while regular bench players may play about 10-25 minutes per game. Of course, these are not hard and fast rules.
    * `PPG`: Points scored per game. The standard for a good scorer is about 20 PPG, and a handful of great scorers may average 30 PPG. For reference, Michael Jordan's *career* average over 15 seasons is about 30 PPG.
    * `RPG`: Rebounds per game. The standard for a good rebounder is about 10 RPG, with season averages above 15 RPG being very unusual today.
    * `APG`: Assists per game. The standard for a good passer is about 6-8 APG. A handful of players consistently average above 10 APG.
    * `SPG`: Steals per game. The top "stealers" may average about 1-2 SPG. At his peak, Magic Johnson averaged about 3 SPG.
    * `BPG`: Blocks per game. The top "blockers" may average about 2-3 BPG. 
    * `FG%`: Overall field goal percentage (shots made divided by shots attempted). Includes 2-pointers and 3-pointers, but not free throws. Most players fall in the 40-55% range, with very efficient or selective players reaching 60%.
    * `2P%`: Two-point shot percentage. As the name implies, does not include 3-pointers. 2P% is usually greater than 3P% (about 45-60%), due to the greater difficulty introduced by the distance of the 3-point line.
    * `3P%`: Three-point shot percentage. 35% is considered average, and 40% is considered good.
    * `FT%`: Free throw shot percentage. 75% is considered average, with excellent shooters reaching up to 90%. On the other hand, poor shooters like Shaquille O'Neal may struggle to reach 50%.
    * `WS`: Career win shares. According to [Basketball-Reference](https://www.basketball-reference.com/about/ws.html), "Win Shares is a player statistic which attempts to divvy up credit for team success to the individuals on the team...it is calculated using player, team and league-wide statistics and the sum of player win shares on a given team will be roughly equal to that team's win total for the season." For reference, Lebron James currently has about 230 career win shares, roughly implying that his contributions amounted to 230 wins.
<p>
* *Normalized Statistics*
    * `PP48`, `RP48`, `AP48`, `SP48`, `BP48`: Variables that we created to understand a player's per-minute performance, extended to 48 minutes. These (very roughly) approximate what players' per-game averages would look like if they always played the entire game, and they were computed by dividing by the minutes per game. For example:
$$\textrm{PP48} = \frac{\textrm{PPG}}{\textrm{MPG}} \times 48 = \frac{\textrm{Total Points}}{\textrm{Total Minutes}} \times 48$$<br>
    * Note: The above statistics are not always reliable, given their inability to account for small sample sizes. A player who only played 1 minute in 1 game but scored 2 points would technically average 96 PP48. Thus, these normalized statistics must always be considered along with the number of games played and the number of minutes per game.
    * `WS/48`: The normalized equivalent of `WS`. For reference, Michael Jordan's career average is about [0.25 WS/48](https://www.basketball-reference.com/leaders/ws_per_48_career.html), roughly implying that he singlehandedly contributed to 25% of a win in every 48 minutes he played.
<p>
* *Ignored Variables*
    * `Bench`: How many repetitions of 185-pound bench presses the player can do. We ignore this because nearly half of the players in our resulting dataset are missing this variable, and because it is likely the [noisiest variable](https://www.sportingnews.com/us/nba/news/nba-draft-combine-test-results-predraft-measurements-bench-press-zero-reps-kevin-durant/1bhegynptq12w1rgn7f7p3yu0w) among all the athletic tests at the combine.

## 2. Setup and Joining

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
player_avgs = pd.read_csv('player_avgs.csv')
nba_combine = pd.read_csv('nba_draft_combine_all_years.csv')
```

We'll rename the `Draft pick` column in `nba_combine` to `Pk`, in order to facilitate joining the two datasets later:

Then, we need to clean the players' names in the `player_avgs` dataframe, so that the part after the slash is not included.

```python
combine_careers_raw = pd.merge(nba_combine, player_avgs, on = ['Player', 'Pk'], how = 'inner')
# combine_careers_raw.to_csv('combine_careers.csv')
combine_careers_raw.head()
```

## 3. Cleaning the Data

Let's look at the structure of our data. We see below that we have 366 rows and 43 columns, but we have quite a few missing values in several columns. There's also seems to be several redundant columns that could be removed (e.g., `Year_x` and `Year_y`).

#### A. Removing Unnecessary Columns

```python
combine_careers = combine_careers_raw.copy()
combine_careers.drop('Unnamed: 0', 1, inplace = True)
```

We also notice that we have two year columns. Are there any discrepancies between the two? The answer is yes:

```python
year_diff = combine_careers[['Year_x', 'Year_y']][combine_careers.Year_x != combine_careers.Year_y]
combine_careers.iloc[year_diff.index][['Player', 'Year_x', 'Year_y']]
```

## 4. Exploratory Data Analysis

#### A. Standardizing Variables

As mentioned in Section 1, Part D, we'll create standardized variables that help us understand players' per-minute production.

```python
cc48 = cc.loc[:,'PPG':'BPG'].apply(lambda x: 48 * (x / cc.MPG))
cc48.columns = ['PP48', 'RP48', 'AP48', 'SP48', 'BP48']
cc = pd.concat([cc, cc48], axis = 1, sort = True)
# comparing per-game to per-48 minutes
cc.loc[:, 'MPG':'BP48'].head()
```

#### A. Examining the Variables: Physical Measurements

Now, we can look at the distributions of individual variables. We'll start with purely physical measurements. As we can see below, most of the physical measurements appear to be bell-shaped and roughly normally distributed, as we might expect. `Body Fat` is an obvious exception, and the hand measurements have some notable outliers. The height variables are somewhat bimodal, with all other variables being mostly unimodal. We'll take a look at each of the unusual features.

```python
physicals = ['Height (No Shoes)', 'Height (With Shoes)', 'Wingspan', 'Standing reach', 
             'Weight', 'Body Fat', 'Hand (Length)', 'Hand (Width)']

fig = plt.figure(figsize = (15, 5))   
for i in range(1, 9):
    plt.subplot(2, 4, i)
    sns.violinplot(x = physicals[i - 1], data = cc[cc[physicals[i - 1]].notnull()], color = 'lightblue')
    plt.subplots_adjust(hspace = 0.5)
    plt.title(physicals[i - 1])
    if physicals[i - 1] == 'Weight':
        plt.xlabel('lb')
    elif physicals[i - 1] == 'Body Fat':
        plt.xlabel('%')
    else:
        plt.xlabel('in')
     
plt.show()
```

First, let's take a closer look at `Body Fat`, specifically the players that have a body fat percentage greater than 13, the upper end of what is expected of most athletes. Unsurprisingly, the players with the highest body fat tend to be picked lower, since teams may view them as "overweight" or "unathletic"; indeed, most of these players were fairly low picks and did not end up playing much in the NBA. The notable exception is DeMarcus Cousins, an All-Star center who dominated possibly because he knew how to use his weight and bulk.

```python
cc[cc['Body Fat'] > 13][['Pk', 'Pos', 'Weight', 'Body Fat', 'G', 'MPG']].sort_values('Body Fat')
```

Second, let's look at `Hand (Length)` and `Hand (Width)`. For `Hand (Length)`, we'll focus on players with the greatest hand lengths, the ones in the 95th percentile. We are interested in looking to see if hand length might be related to having a high RPG. However, we see that RPG varies greatly amongst these players, though RP48 is more stable. In general, there does not appear to be a strong correlation with hand length with the variables shown below.

```python
cc['Hand (Length)'].quantile(q = 0.95)
cc[cc['Hand (Length)'] > 9.5][['Pk','Pos','Hand (Length)','Hand (Width)','MPG','RPG', 'RP48']].sort_values('Hand (Length)')
```

As for `Hand (Width)`, we look at players with the largest hand width and look for any interesting trends. Again, RPG does not seem to be strongly correlated with hand width, and RP48 is not much better. One outlier is Royce White, who famously never played in an NBA game due to his fear of flying. It's also interesting to note the several players that show up on both lists, such as Andrew Nicholson or Kawhi Leonard; despite similar hand measurements, positions, pick range, and even RP48, the latter become a star and the other was largely forgotten.

```python
cc['Hand (Width)'].quantile(q = 0.95)
cc[cc['Hand (Width)'] > 10.5][['Pk','Pos','Hand (Length)','Hand (Width)','MPG','RPG', 'RP48']].sort_values('Hand (Width)')
```

When looking at the violin plot for height without shoes, we notice that the distribution is possibly bimodal. While height may determine whether or not the average person is good at basketball or not, height becomes far less important when our observations are limited to NBA players. This is apparent in how box score statistics such as PPG and RPG vary greatly even amongst the tallest players in the NBA. Indeed, when we look at the in-game performance of the shortest NBA players in this dataset, we see that PPG varies a lot, as certain players excel regardless of their height. Teams seem to understand this, because when we look at picks, being tall does not guarantee that a player will be picked earlier in the draft.

We know that height and wingspan have a strong correlation. Hence, when looking into players with the greatest wingspan in our dataset, we observe similar trends that we say when looking into height (no shoes), which is that box score statistics do not have a strong relationship with wingspan on its own.

```python
cc['Wingspan'].quantile(q = 0.95)
cc[cc['Wingspan'] > 89][['Pk','Pos','Height (No Shoes)', 'Wingspan','G','MPG','PPG','RPG','BPG']].sort_values('Wingspan')
```

## 5. Data Modeling

#### A. Imputation

Since most of `scikit-learn`'s functionality requires our data to have no null values, we will impute the missing values using a variety of methods. First, let's review which columns have missing values.

```python
cc.info()
```

 Kris Dunn is the only player who doesn't have weight/body fat measurements. We'll go ahead and impute the median weight and body fat among guards, since weight/body fat distributions seem slightly skewed from our EDA above.

```python
cc_imp = cc.copy()
cc[cc.Weight.isnull()]
cc_imp.loc[cc.Weight.isnull(), 'Weight'] = cc[cc.Pos == 'G'].Weight.median()
cc_imp.loc[cc['Body Fat'].isnull(), ['Body Fat']] = cc[cc.Pos == 'G']['Body Fat'].median()
```

Now, we'll work on imputing the variables that are missing about 50 observations. We'll use linear regression to "predict" what these missing values may be, based on the other highest-correlated non-missing variable(s). Admittedly, this assumes that these variables have largely linear relatinoships, Starting with the vertical measurements, we'll pick `Standing Reach` and `Body Fat` to predict `Vertical (Max)` and `Vertical (No Step)`, since they're the best predictors that aren't highly correlated.

```python
# we'll use the same method for Vertical (Max) and Vertical (No Step), due to their high correlation
# and we'll use the same method for Vertical (Max Reach) and Vertical (No Step Reach)
cc[['Vertical (Max)', 'Vertical (Max Reach)', 'Vertical (No Step)', 'Vertical (No Step Reach)']].corr()
```

```python
physicals = ['Height (With Shoes)', 'Wingspan', 'Standing reach', 'Weight', 'Body Fat']
cc[physicals + ['Vertical (Max)', 'Vertical (Max Reach)']].corr()
```

```python
from sklearn import linear_model
def impute_linear(impute, regress):
    if type(regress) is not list:
        regress = [regress]
    if type(impute) is not list:
        impute = [impute]
    for colname in impute:
        full = cc_imp[regress + [colname]]
        train = full.dropna()
        test = full[full[colname].isnull()]
        imp = linear_model.LinearRegression()
        imp.fit(X = train.drop(colname, axis = 1), y = train[colname])
        # print(imp.score(X = train.drop(colname, axis = 1), y = train[colname]))
        preds = imp.predict(X = test.drop(colname, axis = 1))
        # impute the linearly regression prediction
        cc_imp.loc[cc[colname].isnull(), colname] = preds
```

We'll repeat the process for `Hand (Length)`, `Hand (Width)`, `Agility`, and `Sprint`. Our choice of variables to use for regressing is shown below, after the tables.

```python
cc[physicals + ['Hand (Length)', 'Hand (Width)']].corr()
```

```python
cc[physicals + ['Agility', 'Sprint']].corr()
```

```python
impute_linear(['Vertical (Max)', 'Vertical (No Step)'], ['Standing reach', 'Body Fat'])
impute_linear(['Vertical (Max Reach)', 'Vertical (No Step Reach)'], 'Wingspan')
impute_linear('Hand (Length)', 'Wingspan')
impute_linear('Hand (Width)', 'Weight')
impute_linear(['Agility', 'Sprint'], 'Standing reach')
```

Now we'll check our dataset again, and it should have much fewer missing values.

```python
cc_imp.info()
```

#### B. Using scikit-learn

Let's create a copy of the imputed dataset and filter it to only include players who have at least played half a season (41 games) and played more than 5 minutes per game. We'll treat all other missing values (in the percentage) as 0, since that's a fair proxy for the impact of a player who didn't attempt any shots.

```python
cc_sk = cc_imp.copy().drop(['Bench', 'Rk','pk_range'], axis = 1)[(cc_imp.G > 40) & (cc_imp.MPG > 5)].fillna(0)
cc_sk.head()
```

```python
combine_sk = cc_sk.loc[:, 'Height (No Shoes)':'Sprint'].assign(Age = cc_sk.Age)
career_sk = cc_sk.loc[:, 'G':]
```

Now, we can try a rudimentary linear regression to predict the five major box score statistics (and their normalized equivalents) and examine the resulting R^2 from each fit. We see that `RP48`, `AP48`, and `BP48` are the most promising.

```python
fig = plt.figure(figsize = (18, 5))   
for i, colname in enumerate(['PPG', 'RPG', 'APG', 'SPG', 'BPG', 'PP48', 'RP48', 'AP48', 'SP48', 'BP48']):
    plt.subplot(2, 5, i + 1)
    var = career_sk[colname]
    per48 = linear_model.LinearRegression()
    per48.fit(X = combine_sk, y = var)
    score = np.round(per48.score(X = combine_sk, y = var), 3)
    # print(metrics.mean_absolute_error(var, per48.predict(combine_sk)))
    plt.plot(per48.predict(combine_sk), var, 'o', c = 'green', alpha = 0.8)
    plt.title('{} | R^2: {}'.format(colname, score))
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.subplots_adjust(hspace = 0.5)
plt.tight_layout()
plt.show()
```

We'll repeat the process for the other variables. The results aren't as promising here.

Let's assess the performance of several models using 5-fold CV. We'll examine 4 different statistics and fit 4 different models to each.

```python
from sklearn.model_selection import LeaveOneOut, cross_val_score, cross_val_predict, train_test_split, cross_validate
from sklearn import tree, metrics, linear_model, ensemble
```

```python
scoring = ['neg_mean_absolute_error', 'neg_mean_squared_error']
models = ['Linear Regression', 'Lasso', 'Decision Tree', 'Random Forest']

def run_model(model, response):
    if model == 'Linear Regression':
        mdl_obj = linear_model.LinearRegression()
    elif model == 'Lasso':
        mdl_obj = linear_model.Lasso(alpha = 0.1)
    elif model == 'Decision Tree':
        mdl_obj = tree.DecisionTreeRegressor()
    elif model == 'Random Forest':
        mdl_obj = ensemble.RandomForestRegressor(n_estimators = 50)
    mdl = mdl_obj.fit(combine_sk, career_sk[response])
    cv_res = pd.DataFrame(cross_validate(mdl, combine_sk, career_sk[response], cv = 5, scoring = scoring))
    return(cv_res.assign(model = model))

def mult_scorer(response, scoring = scoring, models = models):
    # X_train, X_test, y_train, y_test = train_test_split(combine_sk, career_sk[response], test_size = 0.25)
    for i, model in enumerate(models):
        if i == 0:
            mult_res = run_model(model, response)
        else:
            mult_res = pd.concat([mult_res, run_model(model, response)], sort = 'False')                         
    return(mult_res)
```

```python
fig = plt.figure(figsize = (12, 10))   
for i, colname in enumerate(['RP48', 'AP48', 'BP48', 'WS/48']):
    ax = fig.add_subplot(2, 2, i + 1)
    ax.set_ylabel('Test Negative MAE')
    df = mult_scorer(colname, scoring = scoring).pivot(columns = 'model', values = 'test_neg_mean_absolute_error')
    df.index = np.arange(1, len(df) + 1)
    df.index.name = 'CV Run'
    df.plot(title = 'Predicting {}'.format(colname), ax = ax)
    #plt.subplots_adjust(hspace = 0.5)
plt.tight_layout()
plt.show()
```

In final terms, we see that lasso, linear regression, and random forest have similar performance (with random forest the clear winner for RP48). The decision tree's poor performance is probably due to a lack of pruning, as maximum depth was not constrained. Some possible improvements:
* For everything but linear regression, hyperparameter tuning should be conducted via cross-validation. For example, with lasso, we can find a better value of alpha that minimized test MSE. Similarly, for decision trees, we should look for optimal values of max depth, number of leaves, etc.
* We could also perform preprocessing to standardize all our features, to ensure that none of them exert undue influence on our corresponding coefficient estimates (for linear regression and lasso particularly). 
* Or, we could perform feature selection via forward or backward selection, and only create our model based on a subset of the original predictors. This could also reduce test MSE by removing unnecssary predictors and thus reducing variance.

In conclusion, the most promising model for prediction seems to be **random forest**, which already performed near the top of all the models, and which can be tuned via CV for even greater performance gains. The main downside, as shown in the tables above, is that it does take slightly longer to fit and train, making it potentially difficult to scale up to much larger datasets. Otherwise, however, when evaluating all of our models, they can quite accurately predict a number of statistics purely from combine performance (e.g., our average error in predicting RP48 with random forest is under 1.5!). Particularly surprising is the ability to predict AP48 with an average error under 1.4 (also with random forest), which could prove very useful to teams looking for excellent passers in the draft.

At the same time, we shouldn't ignore the usefulness and explainability of linear regression/lasso, which still performs quite well. In the addendum, we'll see that our model assumptions are basically satisfied, which might lead to the linear models being preferred over something more black-box like random forest.
